## 1. 기초 변수 및 자료형 (Basic Variables and Data Types)

파이썬은 동적 타이핑(Dynamic Typing)을 지원하는 언어로, 변수를 선언할 때 자료형을 명시하지 않아도 값이 할당되는 시점에 자동으로 결정됩니다.

### 주요 자료형 및 기능
- **정수형(int) & 실수형(float)**: 수치 데이터를 저장하며 산술 연산에 사용됩니다.
- **문자열(str)**: 텍스트 데이터를 처리합니다. `.upper()`, `.lower()`, `.strip()` 등 다양한 내장 메서드를 통해 데이터를 전처리할 수 있습니다.
- **불리언(bool)**: `True` 또는 `False` 값을 가지며, 주로 조건문과 비교 연산에서 논리적 판단을 위해 사용됩니다.
- **f-string**: 문자열 내부에 변수나 표현식을 직접 삽입하여 가독성 높은 출력을 제공하는 파이썬 3.6+의 기능입니다.

```python
# 1. Variable declaration with basic data types
count = 10          # Integer (int)
price = 19.99       # Float (float)
raw_name = "  python for ai  " # String (str)
is_active = True     # Boolean (bool)

# 2. String manipulation using methods
clean_name = raw_name.strip()      # Removing whitespace
upper_name = clean_name.upper()    # Converting to uppercase
lower_name = clean_name.lower()    # Converting to lowercase

# 3. Boolean logic with comparison operators
threshold = 15.0
is_expensive = price > threshold

# 4. Data output using f-strings
print(f"--- Basic Variable and f-string Demo ---")
print(f"Product Name: '{clean_name}' (Original: '{raw_name}')")
print(f"Uppercase: {upper_name} | Lowercase: {lower_name}")
print(f"Quantity: {count} | Price: ${price}")
print(f"Is the price above ${threshold}? {is_expensive}")
print(f"Total Cost Calculation: ${count * price:.2f}")
print(f"System Active Status: {is_active}")
```

### 2. Python Core Data Structures & Control Flow

Understanding when to use specific data structures is key to efficient Python programming:

| Data Structure | Type | Mutability | Ordering | Key Feature |
| :--- | :--- | :--- | :--- | :--- |
| **List** | `[]` | Mutable | Ordered | Dynamic arrays for collections of items. |
| **Tuple** | `()` | Immutable | Ordered | Fixed sequences (e.g., coordinates, DB records). |
| **Dictionary**| `{}` | Mutable | Unordered* | Key-value pairs for fast lookups. |

*Note: Python 3.7+ dictionaries maintain insertion order.*

In the following section, we combine these structures with **Control Flow** tools:
- **if-elif-else**: Conditional branching.
- **for / range()**: Iterating over sequences or numeric ranges.
- **List Comprehension**: Concise syntax for list creation and filtering.

```python
# 1. Initialize data structures
item_prices = [100, 250, 45, 120, 80]
origin_coords = (37.5665, 126.9780)  # Tuple for fixed location data
product_profile = {
    "id": "A101",
    "name": "Smart Watch",
    "stock": 15,
    "rating": 4.2
}

# 2. if-elif-else logic based on product stock
stock_status = ""
if product_profile['stock'] >= 20:
    stock_status = "High"
elif product_profile['stock'] >= 10:
    stock_status = "Moderate"
else:
    stock_status = "Low"

# 3. for loop with range() to calculate price with 10% tax
taxed_prices = []
for i in range(len(item_prices)):
    taxed_prices.append(item_prices[i] * 1.1)

# 4. List Comprehension: Filter items priced above 100 from taxed_prices
premium_items = [price for price in taxed_prices if price > 100]

# 5. Output results using f-strings
print(f"--- Data Structure & Control Flow Results ---")
print(f"Product: {product_profile['name']} | Stock Level: {stock_status} ({product_profile['stock']} units)")
print(f"Coordinates: Latitude {origin_coords[0]}, Longitude {origin_coords[1]}")
print(f"Original Prices: {item_prices}")
print(f"Taxed Prices (10%): {[round(p, 2) for p in taxed_prices]}")
print(f"Premium Items (>100): {[round(p, 2) for p in premium_items]}")
```

## 3. Functions and Code Reusability

Functions are self-contained blocks of code that perform a specific task. They are the building blocks of reusable and modular code in Python.

### Key Components of a Function:
- **`def` Keyword**: Used to define a new function followed by the function name.
- **Parameters**: Inputs passed into the function inside the parentheses to customize its behavior.
- **Return Value (`return`)**: The data sent back from the function to the caller.
- **Docstrings**: String literals enclosed in triple quotes (`"""`) immediately following the function definition. They provide a standardized way to document what the function does, its parameters, and its return type, which is essential for maintainable code.

```python
def get_list_statistics(numbers):
    """
    Calculates the total sum, average, and maximum value of a list.

    Args:
        numbers (list): A list of numerical values (int or float).

    Returns:
        dict: A dictionary containing 'total', 'average', and 'max_value'.
    """
    if not numbers:
        return {"total": 0, "average": 0, "max_value": 0}

    total = sum(numbers)
    average = total / len(numbers)
    max_value = max(numbers)

    return {
        "total": total,
        "average": average,
        "max_value": max_value
    }

# Call the function with the existing item_prices list
stats = get_list_statistics(item_prices)

# Output the results using f-string
print(f"--- List Statistics Summary ---")
print(f"Target List: {item_prices}")
print(f"Total Sum: {stats['total']}")
print(f"Average Price: {stats['average']:.2f}")
print(f"Maximum Price: {stats['max_value']}")
```

### 4. Advanced Data Analysis and Visualization

In this section, we transition from core Python to specialized libraries for AI: **NumPy**, **Pandas**, and **Matplotlib**.

- **NumPy**: Essential for numerical computing and vectorization.
- **Pandas**: The industry standard for structured data manipulation using DataFrames.
- **Matplotlib**: A comprehensive library for creating static, animated, and interactive visualizations.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 1. NumPy: Calculate statistics from item_prices
# item_prices: [100, 250, 45, 120, 80]
np_prices = np.array(item_prices)
price_mean = np.mean(np_prices)
price_std = np.std(np_prices)
price_median = np.median(np_prices)

print(f"--- NumPy Statistical Analysis ---")
print(f"Mean Price: {price_mean:.2f}")
print(f"Standard Deviation: {price_std:.2f}")
print(f"Median Price: {price_median:.2f}\n")

# 2. Pandas: Create DataFrame and show summary statistics
data = {
    'Product': ['Item 1', 'Item 2', 'Item 3', 'Item 4', 'Item 5'],
    'Price': item_prices,
    'Rating': [4.5, 3.8, 4.2, 4.8, 4.0]
}
df_products = pd.DataFrame(data)

print("--- Pandas Descriptive Statistics ---")
print(df_products.describe())

# 3 & 4. Matplotlib: Visualize and enhance
plt.figure(figsize=(10, 6))
plt.bar(df_products['Product'], df_products['Price'], color='teal', label='Price per Unit')

# Adding title and labels
plt.title('Product Price Comparison', fontsize=14)
plt.xlabel('Product Name', fontsize=12)
plt.ylabel('Price ($)', fontsize=12)

# Visual enhancements
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()

# Display the plot
plt.show()
```

### 5. 객체지향 프로그래밍(OOP) 및 예외 처리 기초

**객체지향 프로그래밍(Object-Oriented Programming, OOP)**은 프로그램을 '객체(Object)'들의 집합으로 구성하는 패러다임입니다. 이를 통해 코드의 재사용성과 유지보수성을 높일 수 있습니다.

#### 핵심 개념:
- **클래스(Class)**: 객체를 만들기 위한 **설계도** 또는 청사진입니다.
- **속성(Attribute)**: 객체가 가지는 **상태**나 특징을 나타내는 변수입니다. (예: 학생의 이름, 점수)
- **메서드(Method)**: 객체가 수행하는 **행위**나 기능을 나타내는 함수입니다. (예: 평균 계산하기, 정보 출력하기)

#### 예외 처리(Exception Handling):
프로그램 실행 중 발생할 수 있는 오류를 미리 대비하는 기법입니다. `try-except-finally` 구문을 사용하면 예상치 못한 에러가 발생해도 프로그램이 갑자기 종료되지 않고 안전하게 처리할 수 있습니다.

```python
class DataAnalyzer:
    def __init__(self, data_name):
        """
        Initializes the analyzer with a specific name.
        """
        self.data_name = data_name
        self.data = []

    def load_data(self, data_list):
        self.data = data_list

    def calculate_statistics(self):
        """
        Calculates statistics with try-except-finally for robust error handling.
        """
        print(f"--- Analyzing: {self.data_name} ---")
        try:
            if not self.data:
                raise ValueError("The data list is empty.")

            total = sum(self.data)
            count = len(self.data)
            average = total / count

            print(f"Success: Average is {average:.2f}")
            return average

        except ZeroDivisionError:
            print("Error: Division by zero occurred.")
        except TypeError:
            print("Error: Data contains non-numeric values.")
        except ValueError as e:
            print(f"Error: {e}")
        except Exception as e:
            print(f"An unexpected error occurred: {e}")
        finally:
            print(f"Analysis attempt for '{self.data_name}' is finished.\n")

# 1. Normal data test
clean_analyzer = DataAnalyzer("Normal Scores")
clean_analyzer.load_data([85, 90, 78, 92])
clean_analyzer.calculate_statistics()

# 2. Empty list test (ValueError)
empty_analyzer = DataAnalyzer("Empty Dataset")
empty_analyzer.load_data([])
empty_analyzer.calculate_statistics()

# 3. Invalid data type test (TypeError)
mixed_analyzer = DataAnalyzer("Mixed Dataset")
mixed_analyzer.load_data([10, 'twenty', 30])
mixed_analyzer.calculate_statistics()
```

## 6. System Interaction and Parallel Computing

To build efficient AI pipelines, it is often necessary to interact with the underlying operating system or handle heavy computational tasks in parallel.

### Subprocess Module
- **Purpose**: Used to run **external applications** or shell commands (e.g., `ls`, `git`, or compiled C++ binaries).
- **Mechanism**: Spawns a new process independent of the current Python interpreter.
- **Use Case**: Automating system tasks or using tools not natively written in Python.

### Multiprocessing Module
- **Purpose**: Used to run **Python functions** in parallel to speed up CPU-bound tasks.
- **Mechanism**: Spawns multiple Python processes, each with its own memory space, effectively bypassing the **Global Interpreter Lock (GIL)**.
- **Use Case**: Large-scale data preprocessing, image transformations, or training ensemble models.

```python
import subprocess
import multiprocessing
import time

# 1. Subprocess: Execute a system command and capture output
try:
    print("--- Executing 'ls -l' via Subprocess ---")
    result = subprocess.run(['ls', '-l', '/content/sample_data'], capture_output=True, text=True, check=True)
    print(f"Output preview:\n{result.stdout.splitlines()[0] if result.stdout else 'Empty'}\n")
except subprocess.CalledProcessError as e:
    print(f"Subprocess Error: {e}")

# 2. Multiprocessing: Define CPU-intensive task
def cpu_intensive_task(n):
    return sum(i * i for i in range(n))

if __name__ == '__main__':
    # List of large inputs
    tasks = [10**7] * 4

    print(f"--- Starting Parallel Processing ({len(tasks)} tasks) ---")
    start_time = time.time()

    # Use Pool to map the function across inputs concurrently
    with multiprocessing.Pool(processes=multiprocessing.cpu_count()) as pool:
        results = pool.map(cpu_intensive_task, tasks)

    end_time = time.time()

    print(f"Parallel Execution Time: {end_time - start_time:.4f} seconds")
    print(f"Results summary: {len(results)} items computed.")
```

## 7. Lazy Evaluation with Generators

**Lazy evaluation** is an evaluation strategy which delays the evaluation of an expression until its value is actually needed. In Python, this is primarily implemented using **Generators**.

### Why use Lazy Evaluation in AI?
- **Memory Efficiency**: You can process datasets larger than your RAM because only one item is stored at a time.
- **Performance**: It avoids unnecessary computations if you only need to process a part of the data.

### Key Tools:
- **`yield` keyword**: Turns a function into a generator.
- **Generator Expressions**: Similar to list comprehensions but use parentheses `()` instead of square brackets `[]`.

```python
import sys

# 1. Standard List (Eager Evaluation)
eager_list = [i for i in range(10000)]
print(f"Eager List Size: {sys.getsizeof(eager_list)} bytes")

# 2. Generator (Lazy Evaluation)
lazy_gen = (i for i in range(10000))
print(f"Lazy Generator Size: {sys.getsizeof(lazy_gen)} bytes")

# 3. Generator Function using 'yield'
def lazy_data_loader(n):
    print("--- Starting data stream ---")
    for i in range(n):
        # Logic: Only compute/load when requested
        yield f"Processed Item {i}"

# Usage
stream = lazy_data_loader(3)
print("Generator object created, but no items processed yet.")

print(next(stream)) # First item is evaluated here
print(next(stream)) # Second item is evaluated here

print("--- Iterating through remaining ---")
for item in stream:
    print(item)
```

** Lazy Evaluation **
- Lazy evaluation is a powerful concept in Python, especially when dealing with large datasets in AI. It allows you to process items one by one instead of loading everything into memory at once. I'll add an explanation and a code example for you.
The execution results demonstrate a significant difference in memory usage: the eager list consumed 85,176 bytes, while the lazy generator used only 192 bytes for the same 10,000 items. Additionally, you can see that the 'Starting data stream' message only appeared when we actually requested the first item using next(), proving that the computation was deferred until needed.
