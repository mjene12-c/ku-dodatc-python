### 3. JAX vs NumPy
JAX는 고성능 머신러닝 연구를 위해 구글에서 개발한 라이브러리로, GPU/TPU 가속과 자동 미분(AutoGrad) 기능이 내장된 **'강력한 버전의 NumPy'** 입니다.
`jax.numpy`의 문법은 기존 `numpy`와 99% 일치하지만 더 강력합니다.

```python
# 주의: 환경에 따라 jax 모듈이 설치되어 있어야 실행됩니다.
# !pip install jax jaxlib
try:
    import jax.numpy as jnp
    from jax import grad
    
    # JAX를 이용한 배열 생성 (문법이 numpy와 동일)
    j_arr = jnp.array([10., 20., 30.])
    print("JAX 배열 (DeviceArray):", j_arr)
    print("JAX 연산 (j_arr * 2) =", j_arr * 2)
    
    # 자동미분 맛보기: 함수 f(x) = x^2
    def f(x):
        return x**2
    
    # f(x)의 도함수 생성 (2x)
    df = grad(f)
    
    print("\nf(3.0) =", f(3.0))         # 9.0
    print("f'(3.0) [미분값] =", df(3.0)) # 6.0
    
except ImportError:
    print("현재 환경에 jax가 설치되어 있지 않습니다. `pip install jax jaxlib`을 통해 설치 후 실행해보세요.")
```

### 5. [심화] 고윳값(Eigenvalue)과 특이값 분해(SVD)

유사 역행렬(Pseudo Inverse)을 이해하기 위한 필수 뼈대 지식입니다.\n
데이터의 주요 축을 찾거나(PCA 차원 축소) 직사각형 행렬을 다루기 위해 **고윳값 분해(Eigendecomposition)**와 **특이값 분해(SVD)**를 학습합니다.

```python
import numpy as np

# -----------------------------------
# 1. 고윳값(Eigenvalue)과 고유벡터(Eigenvector)
# -----------------------------------
# (정방행렬 2x2)
A_sq = np.array([
    [4, 1],
    [2, 3]
])

print("=== 정방행렬 A_sq ===\n", A_sq)

# 고윳값(eigenvalues)과 고유벡터(eigenvectors) 계산
eigenvalues, eigenvectors = np.linalg.eig(A_sq)

print("\n고윳값 (Eigenvalues):", eigenvalues)
print("고유벡터 (Eigenvectors):\n", eigenvectors)

# 검증: A * v = lambda * v
v0 = eigenvectors[:, 0]    # 첫번째 고유벡터
lambda0 = eigenvalues[0]   # 첫번째 고윳값

print("\n검증 (A * v0):", np.dot(A_sq, v0))
print("검증 (lambda0 * v0):", lambda0 * v0)
print("(두 결과가 같으면 수학적으로 올바른 것입니다!)")
```

```python
# -----------------------------------
# 2. 특이값 분해 (Singular Value Decomposition, SVD)
# -----------------------------------
# (직사각형 행렬 3x2 - 앞선 데이터 예시)
A_rect = np.array([
    [1, 2],
    [3, 4],
    [5, 6]
])

print("=== 직사각형 행렬 A_rect (3x2) ===\n", A_rect)

# 특이값 분해 (A = U * Sigma * V^T)
# U: (3x3) 왼쪽 특이벡터 행렬
# S: (2,) 1차원 배열로 된 특이값 (Singular values)
# VT: (2x2) 오른쪽 특이벡터 행렬
U, S, VT = np.linalg.svd(A_rect)

print("\nU 행렬 (행 차원 회전):\n", U)
print("\n특이값 S (중요도/크기, 내림차순):", S) # S 데이터가 분산/정보량을 대변합니다
print("\nVT 행렬 (열 차원 회전):\n", VT)
```

### 6. [심화] 유사 역행렬 (Pseudo Inverse)

정방행렬(N x N)이 아닌 행렬에 대해서도 역행렬과 비슷한 역할을 하는 행렬을 구하는 방법입니다.\n
머신러닝의 기초 알고리즘 중 평범한 '선형 회귀(Linear Regression)'의 최적 파라미터(가중치)를 오차 없이 단번에 계산할 때 사용됩니다.\n

NumPy에서는 `np.linalg.pinv(A)` 함수를 사용하여 구할 수 있습니다.

```python
import numpy as np

# 1. 3행 2열의 직사각형 행렬 A (정방행렬이 아님)
A = np.array([
    [1, 2],
    [3, 4],
    [5, 6]
])

print("원본 행렬 A (3x2):\n", A)

# 2. 유사 역행렬 계산 (Moore-Penrose Pseudo Inverse)
A_pinv = np.linalg.pinv(A)

print("\n유사 역행렬 A_pinv (2x3):\n", A_pinv)

# 3. 유사 역행렬의 성질 확인 (A_pinv * A = I, 항등행렬)
# 완벽한 항등행렬이 나오는지 확인해봅시다. (소수점 아래 미세 오차 무시)
identity_approx = np.dot(A_pinv, A)
print("\nA_pinv와 A의 내적 (근사적인 단위행렬):\n", np.round(identity_approx, 2))
```

### 7. [실무 응용] SVD와 Pseudo Inverse를 활용한 GPS 위치 계산

우리가 매일 사용하는 스마트폰의 GPS는 위성(Satellite)으로부터 **위성의 현재 3차원 위치**와 **신호 송신 시간($t_i$)** 데이터를 받습니다.\n
수신기는 자신의 시계로 **수신 시간($t_{rec}$)**을 측정하여 위성과의 거리(Pseudo-range)를 계산하지만, 스마트폰의 시계는 부정확하기 때문에 **수신기 시계 오차(b)**가 미지수로 추가됩니다.\n

우리가 구해야 할 미지수는 4개($x, y, z, b$)인데, 상용 GPS는 6개 이상의 위성 신호를 잡습니다. 즉, 방정식이 미지수보다 많은 **초과 결정 시스템(Overdetermined System)**이 되므로 역행렬을 직접 구할 수 없습니다.\n
이 때 SVD를 이용한 **유사 역행렬(Pseudo Inverse)**로 가장 오차가 적은 최적의 위치(x, y, z)를 한 번에 계산해냅니다!

```python
import numpy as np

# 빛의 속도 (km/s)
C = 299792.458 

# [1. 시뮬레이션 환경 구성]
# 6개 위성의 임의 3차원 좌표 (고도 약 20,000km 수준)
satellites = np.array([
    [15000, 12000, 20000],
    [-10000, 18000, 22000],
    [1000, -15000, 21000],
    [14000, -10000, 23000],
    [-12000, -8000, 24000],
    [-5000, 16000, 25000]
])

# 사용자의 실제 3차원 위치 (우리가 SVD 역산으로 찾아내야 할 정답)
true_pos = np.array([300, 400, 50])
true_clock_bias = 0.00001 # 수신기의 실제 시간 오차 (초 단위)

# (가상 생성) 위성이 신호를 보낸 시간 t_i 배열 생성
# 실제 거리 / 빛의 속도 = 도달에 걸린 순수 시간
transmit_times = []
receiver_time = 100.0 # 수신기가 신호를 받은 수신기 기준 시간(불안정함)

# 각 위성이 신호를 송신한 시간(t_i)을 역으로 시뮬레이션하여 쏜다고 가정
for sat in satellites:
    real_dist = np.linalg.norm(sat - true_pos)
    pure_travel_time = real_dist / C
    
    # 수신기가 100.0초에 받았다고 할때, 실제 세상 시간은 100.0 - 오차
    true_receiver_time = receiver_time - true_clock_bias
    t_i = true_receiver_time - pure_travel_time
    transmit_times.append(t_i)
transmit_times = np.array(transmit_times)

# --- 여기까지가 GPS 위성 시스템이 허공에 쏜 "숨겨진 정답" 입니다 ---
```

```python
# [2. 수신기가 데이터를 받아서 Pseudo-range (의사 거리) 측정]
# 수신기는 위성 위치(satellites)와 송신 시간(transmit_times)을 전달받고, 자신의 수신 시간(receiver_time)을 뺍니다.

pseudo_ranges = []
for t_i in transmit_times:
    # P = c * (수신기 시계 시간 - 위성 송신 시간)
    # 이 거리 안에는 (실제 거리 + c*수신기 시간오차) 가 섞여 있습니다.
    noise = np.random.normal(0, 0.001) # 대기권 통과 전파 지연 등 약간의 노이즈
    P_i = C * (receiver_time - t_i) + noise 
    pseudo_ranges.append(P_i)
pseudo_ranges = np.array(pseudo_ranges)
```

```python
# [3. 선형 방정식 A * x = b 세우기]
# 비선형 모델(구의 방정식)을 위성 1번 식을 뺌으로써 연립 일차 방정식(선형 모델)으로 변환
# 우리가 구할 미지수 배열 x_vec = [x, y, z, C*시계오차]
A = []
b = []

x1, y1, z1 = satellites[0]
P1 = pseudo_ranges[0]

for i in range(1, 6):
    xi, yi, zi = satellites[i]
    Pi = pseudo_ranges[i]
    
    # A 행렬: [ 2(Xi - X1), 2(Yi - Y1), 2(Zi - Z1), -2(Pi - P1) ]
    A_row = [
        2 * (xi - x1),
        2 * (yi - y1),
        2 * (zi - z1),
        -2 * (Pi - P1) # c*d 의 계수
    ]
    A.append(A_row)
    
    # b 벡터: Xi^2 - X1^2 + Yi^2 - Y1^2 + Zi^2 - Z1^2 - Pi^2 + P1^2
    b_val = (xi**2 - x1**2) + (yi**2 - y1**2) + (zi**2 - z1**2) - (Pi**2 - P1**2)
    b.append(b_val)

A = np.array(A) # 크기: 5 x 4
b = np.array(b) # 크기: 5
```

```python
# [4. 유사 역행렬(Pseudo Inverse)로 해 구하기]
# A는 5x4의 직사각형 행렬이므로 역행렬(A^-1)이 없습니다.
# np.linalg.pinv(A) 를 통해 SVD 기반의 최적해를 도출합니다!
A_pinv = np.linalg.pinv(A)

# 행렬의 내적 연산 (A_pinv * b) 로 x_vec (x, y, z, c*d) 도출
calculated_vec = np.dot(A_pinv, b)

calculated_pos = calculated_vec[:3]
calculated_clock_bias = calculated_vec[3] / C # 오차 거리를 다시 시간(초)으로 복원

print("=== 실제 정답 위치 및 오차 ===")
print(f"X: {true_pos[0]:.2f}, Y: {true_pos[1]:.2f}, Z: {true_pos[2]:.2f}")
print(f"Clock Bias: {true_clock_bias:.6f} 초")

print("\n=== SVD 유사 역행렬로 역산출 한 결과 ===")
print(f"X: {calculated_pos[0]:.2f}, Y: {calculated_pos[1]:.2f}, Z: {calculated_pos[2]:.2f}")
print(f"Clock Bias: {calculated_clock_bias:.6f} 초")

print("\n[결론] 직사각형 행렬 시스템임에도 SVD 덕분에 오차를 최소화한 수신기의 정확한 좌표와 시간까지 알아냈습니다!")
```
